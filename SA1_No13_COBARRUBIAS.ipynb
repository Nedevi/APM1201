```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

df = pd.DataFrame({'List Size': list_size, 'Execution Time': t})

plt.figure(figsize=(10, 6))
sns.lineplot(x='List Size', y='Execution Time', data=df)

plt.title('Execution Time vs List Size')
plt.xlabel('List Size')
plt.ylabel('Execution Time (seconds)')

plt.grid(True)
plt.xscale('log')
plt.yscale('log')

plt.show()
```
