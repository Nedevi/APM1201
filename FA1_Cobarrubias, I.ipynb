```python
def is_multiple(n,m):
    if (m%n == 0):
        return True
    else:
        return False
        
print(is_multiple(2,5))
print(is_multiple(2,4))
```

```python
import itertools

chars = ['c', 'a', 't', 'd', 'o', 'g']
n = 1
while n < (len(chars) + 1):
    permutations = itertools.permutations(chars, n)
    for p in permutations:
        print(''.join(p))
    n+=1
```
