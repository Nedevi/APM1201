```python
##3.55
import time
import random

def prefix_average1(S):
    n = len(S)
    A = [0] * n
    for j in range(n):
        total = 0 # begin computing S[0] + ... + S[j]
        for i in range(j + 1):
            total += S[i]
        A[j] = total / (j+1) # record the average
    return A

def prefix_average2(S):
    n = len(S)
    A = [0] * n
    for j in range(n):
        A[j] = sum(S[0 : j + 1])/(j + 1)
    return A

def prefix_average3(S):
    n = len(S)
    A = [0] * n
    total = 0
    for j in range(n):
        total += S[j]
        A[j] = total / (j + 1)
    return A

S = []
time1 = []
time2 = []
time3 = []
numval = []
for i in range(1000):
    numval.append(i)
    n = random.randint(0,100)
    S.append(n)
    now = time.time_ns()
    prefix_average1(S)
    time1.append(time.time_ns() - now)
    now = time.time_ns()
    prefix_average2(S)
    time2.append(time.time_ns() - now)
    now = time.time_ns()
    prefix_average3(S)
    time3.append(time.time_ns() - now)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

df = pd.DataFrame()
df['Number of Values'] = numval
df['Time of Ave1 (ns)'] = time1
df['Time of Ave2 (ns)'] = time2
df['Time of Ave3 (ns)'] = time3



plt.figure(figsize=(10, 6))

sns.scatterplot(data=df, x='Number of Values', y='Time of Ave1 (ns)', color='pink', label='Prefix_Average1')
sns.scatterplot(data=df, x='Number of Values', y='Time of Ave2 (ns)', color='green', label='Prefix_Average2')
sns.scatterplot(data=df, x='Number of Values', y='Time of Ave3 (ns)', color='red', label='Prefix_Average3')

plt.xscale('symlog')
plt.yscale('symlog')
plt.xlabel('Number of Values')
plt.ylabel('Time (ns)')
plt.legend()

plt.show()
```
